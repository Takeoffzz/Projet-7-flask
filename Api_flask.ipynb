```python
dataframe = pd.read_csv('app_train.csv')
```

```python
col = 'TARGET'
X = dataframe.loc[:, dataframe.columns != col]
y = dataframe['TARGET']
```

```python
clf = load('LGBM_TUNED.joblib') 
```

```python
clf.fit(X, y)
```

```python
def recuperer_score_par_ID(id):
        df2 = X[X['SK_ID_CURR'] == id]
        y_pred = clf.predict_proba(df2)
        return y_pred
```

```python
id = widgets.Text(placeholder='ID du client ici', type="number")
```

```python
seuil = 0.478
```

```python
# button send

button_send = widgets.Button(
                description='Envoyer',
                tooltip='Send',
                style={'description_width': 'initial'}
            )

output = widgets.Output()

def on_button_clicked(event):
    with output:
        clear_output()
        print("Sent message: ")
        if id == dataframe['SK_ID_CURR']:
            recuperer_score_par_ID(id)
            score = y_pred
            if score > seuil :
                print(f"Score du client : {score}, le crédit peut-être accordé.")
            else :
                print(f"Score du client : {score}, ne pas accorder de crédit.")
        else:
            print(f"Mauvais ID, pas de client renseigné")
        
        
button_send.on_click(on_button_clicked)

vbox_result = widgets.VBox([button_send, output])
```
